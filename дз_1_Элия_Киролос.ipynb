##ğŸ§  Ğ§Ğ°ÑÑ‚ÑŒ 3 â€” ĞĞ¿Ñ‚Ğ¸Ğ¼Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ Ñ‚Ğ¸Ğ¿Ğ¾Ğ² Ğ² pandas: Housing

###ğŸ“¥ Ğ§Ñ‚ĞµĞ½Ğ¸Ğµ Ğ¸ Ğ¿Ğ°Ğ¼ÑÑ‚ÑŒ Ğ´Ğ¾

```python
import pandas as pd, numpy as np, warnings
df = pd.read_csv("/content/Housing.csv")
base_mem_mb = df.memory_usage(deep=True).sum()/(1024**2)
pd.DataFrame({"metric":["memory_before_mb"],"value":[base_mem_mb]})
```

###ğŸ§­ ĞŸĞ¾Ğ´Ğ±Ğ¾Ñ€ Ğ¾Ğ¿Ñ‚Ğ¸Ğ¼Ğ°Ğ»ÑŒĞ½Ñ‹Ñ… Ñ‚Ğ¸Ğ¿Ğ¾Ğ²

```python
from pandas.api.types import is_integer_dtype, is_float_dtype, is_bool_dtype, is_object_dtype

def suggest_dtype(s):
    non_null = s.notna().sum()
    nunique = s.nunique(dropna=True)
    pct_unique = (nunique / non_null) if non_null else np.nan
    orig = str(s.dtype)
    sugg = orig
    reason = "keep"
    if is_integer_dtype(s):
        s2 = pd.to_numeric(s, downcast="integer")
        sugg = str(s2.dtype)
        reason = "downcast_integer"
    elif is_float_dtype(s):
        s2 = pd.to_numeric(s, downcast="float")
        sugg = str(s2.dtype)
        reason = "downcast_float"
    elif is_bool_dtype(s):
        sugg = "boolean" if s.isna().any() else "bool"
        reason = "bool"
    elif is_object_dtype(s):
        with warnings.catch_warnings():
            warnings.filterwarnings("ignore", message="Could not infer format, so each element will be parsed individually.*")
            s_dt = pd.to_datetime(s, errors="coerce", cache=True)
        ok = s_dt.notna().sum()
        if non_null and ok / non_null >= 0.95 and ok > 0:
            sugg = "datetime64[ns]"
            reason = "datetime"
        else:
            vals = s.dropna().astype("string").str.lower().str.strip()
            m_true = {"true","yes","y","1"}
            m_false = {"false","no","n","0"}
            if non_null and (vals.isin(m_true.union(m_false)).mean() >= 0.95):
                sugg = "boolean"
                reason = "bool_from_text"
            else:
                if non_null and nunique > 0 and (nunique <= max(50, int(0.5*non_null))):
                    sugg = "category"
                    reason = "low_cardinality"
                else:
                    sugg = "object"
                    reason = "text"
    return orig, sugg, non_null, nunique, pct_unique, reason

profile = []
for c in df.columns:
    profile.append(suggest_dtype(df[c])[:0]+(c,)+suggest_dtype(df[c]))
report = pd.DataFrame([{"column": c, "orig_dtype": o, "suggested_dtype": s, "non_null": nn, "n_unique": nu, "pct_unique": pu, "reason": r}
                       for (c,(o,s,nn,nu,pu,r)) in zip(df.columns, [suggest_dtype(df[c]) for c in df.columns])])
report
```

###ğŸ”„ ĞŸÑ€Ğ¸Ğ¼ĞµĞ½ĞµĞ½Ğ¸Ğµ Ñ‚Ğ¸Ğ¿Ğ¾Ğ²

```python
df_opt = df.copy()
bool_map = {"true": True, "yes": True, "y": True, "1": True, "false": False, "no": False, "n": False, "0": False}
for _, row in report.iterrows():
    c = row["column"]
    target = row["suggested_dtype"]
    if target == row["orig_dtype"]:
        continue
    if target.startswith("int") or target.startswith("uint"):
        df_opt[c] = pd.to_numeric(df_opt[c], downcast="integer")
    elif target.startswith("float"):
        df_opt[c] = pd.to_numeric(df_opt[c], downcast="float")
    elif target == "boolean":
        vals = df_opt[c].astype("string").str.lower().str.strip()
        df_opt[c] = vals.map(bool_map).astype("boolean")
    elif target == "bool":
        df_opt[c] = df_opt[c].astype(bool)
    elif target == "datetime64[ns]":
        with warnings.catch_warnings():
            warnings.filterwarnings("ignore", message="Could not infer format, so each element will be parsed individually.*")
            df_opt[c] = pd.to_datetime(df_opt[c], errors="coerce", cache=True)
    elif target == "category":
        df_opt[c] = df_opt[c].astype("category")
```

###âš–ï¸ ĞŸĞ°Ğ¼ÑÑ‚ÑŒ Ğ¿Ğ¾ÑĞ»Ğµ Ğ¸ ÑĞºĞ¾Ğ½Ğ¾Ğ¼Ğ¸Ñ

```python
opt_mem_mb = df_opt.memory_usage(deep=True).sum()/(1024**2)
pd.DataFrame({"metric":["memory_before_mb","memory_after_mb","saving_mb","saving_pct"],
              "value":[base_mem_mb, opt_mem_mb, base_mem_mb-opt_mem_mb, (1-opt_mem_mb/base_mem_mb)*100]})
```

###ğŸ—‚ï¸ Ğ˜Ğ·Ğ¼ĞµĞ½ĞµĞ½Ğ¸Ñ Ğ¿Ğ¾ ÑÑ‚Ğ¾Ğ»Ğ±Ñ†Ğ°Ğ¼

```python
mem_before = df.memory_usage(index=False, deep=True)
mem_after = df_opt.memory_usage(index=False, deep=True)
delta = (mem_before - mem_after).sort_values(ascending=False)
changes = report.loc[report["orig_dtype"] != report["suggested_dtype"], ["column","orig_dtype","suggested_dtype","reason"]].set_index("column")
changes = changes.join(pd.DataFrame({"mem_before":mem_before, "mem_after":mem_after, "mem_saved":delta}))
changes.sort_values("mem_saved", ascending=False)
```
